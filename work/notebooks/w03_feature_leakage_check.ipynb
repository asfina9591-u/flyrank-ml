## 1. Build the feature vector

*Code that actually builds it — engineered features, categorical handling, fills.*

```python
import os
import subprocess
import pandas as pd

REPO_URL = "https://github.com/asfina9591-u/flyrank-ml"
REPO_DIR = "flyrank-ml"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git", "clone", REPO_URL], check=True)

os.chdir(REPO_DIR)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Selected features
features = [
    "search_volume",
    "competition",
    "ctr",
    "avg_position",
    "engagement_rate",
    "word_count"
]

# Fill missing values
X = df[features].fillna(df[features].median(numeric_only=True))

print("Feature Vector Shape:", X.shape)
display(X.head())
```

## 3. The leakage hunt

*Attack your own features: label-derived columns, future windows, product flags. Show the test.*

```python
print("Checking for possible leakage columns...\n")

possible_leakage = [
    "trend_direction",
    "trend_pct"
]

for col in possible_leakage:
    if col in df.columns:
        print(f"Potential leakage column found: {col}")

print("\nThese columns are excluded from the feature vector because they contain the outcome we are trying to predict or are directly related to it.")
```
