## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import os
import subprocess
import pandas as pd

REPO_URL = "https://github.com/asfina9591-u/flyrank-ml"
REPO_DIR = "flyrank-ml"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git","clone",REPO_URL],check=True)

os.chdir(REPO_DIR)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Fill missing values
df["ctr"] = df["ctr"].fillna(df["ctr"].median())
df["engagement_rate"] = df["engagement_rate"].fillna(df["engagement_rate"].median())
df["avg_position"] = df["avg_position"].fillna(df["avg_position"].median())

# Simple baseline score
df["baseline_score"] = (
    (1 - df["ctr"]) * 0.4 +
    (1 - df["engagement_rate"] / 100) * 0.3 +
    (df["avg_position"] / df["avg_position"].max()) * 0.3
)

baseline = df.sort_values("baseline_score", ascending=False)

os.makedirs("work/outputs", exist_ok=True)

baseline.to_csv("work/outputs/baseline_action_score.csv", index=False)

print("Baseline queue created.")
display(baseline.head())
```

## 3. Top-20 review

*For each of the top 20: action, reason code, confidence note, and what would make it wrong.*

```python
top20 = baseline.head(20).copy()

top20["Recommended Action"] = "Refresh Content"
top20["Reason Code"] = "RC1 + RC2 + RC3"
top20["Confidence"] = "Medium"
top20["Possible Error"] = "May miss pages affected by factors not captured in the dataset."

display(top20[
    [
        "content_id",
        "baseline_score",
        "Recommended Action",
        "Reason Code",
        "Confidence",
        "Possible Error"
    ]
])
```
