## 3. Train + compare vs my baseline

*Same data, same metric, same split as your Week-4 baseline. Show the table.*

```python
import os
import subprocess
import json
import pandas as pd

REPO_URL = "https://github.com/asfina9591-u/flyrank-ml"
REPO_DIR = "flyrank-ml"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git","clone",REPO_URL],check=True)

os.chdir(REPO_DIR)

# Run the provided ML pipeline
subprocess.run(["python","scripts/run_all.py"])

# Read results
with open("outputs/model_results.json") as f:
    results = json.load(f)

baseline = results["baseline"]["baseline_precision_at_50"]
rf = results["models"]["random_forest"]["precision_at_50"]

comparison = pd.DataFrame({
    "Model":["Baseline Rule","Random Forest"],
    "Precision@50":[baseline,rf]
})

display(comparison)
```
