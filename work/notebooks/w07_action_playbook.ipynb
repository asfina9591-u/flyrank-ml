## 5. Exports for the paper

*Write the queue (and any figures you want to reuse) to work/outputs/ — your paper builds on these files.*

```python
import os
import subprocess
import shutil

REPO_URL = "https://github.com/asfina9591-u/flyrank-ml"
REPO_DIR = "flyrank-ml"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git", "clone", REPO_URL], check=True)

os.chdir(REPO_DIR)

# Run the pipeline to generate outputs
subprocess.run(["python", "scripts/run_all.py"])

os.makedirs("work/outputs", exist_ok=True)

# Copy refresh queue if available
if os.path.exists("outputs/refresh_queue.csv"):
    shutil.copy(
        "outputs/refresh_queue.csv",
        "work/outputs/refresh_queue.csv"
    )

print("Export completed successfully.")
print("Queue saved to work/outputs/")
```
